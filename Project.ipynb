```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the Match_summary sheet from the Excel file
df_match_summary = pd.read_excel('ICC_T20_WC_2022.xlsx', sheet_name='Match_summary')

# Filter the data for qualifier matches
df_qualifier_matches = df_match_summary[df_match_summary['Match Type'] == 'Qualifier']

# Calculate the frequency of each winner
winner_counts = df_qualifier_matches['winner'].value_counts()

# Create a bar plot using Seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x=winner_counts.index, y=winner_counts.values, palette='Set3')
plt.xlabel('Team')
plt.ylabel('Frequency')
plt.title('Qualifier Matches: Winners and Frequencies')
plt.xticks(rotation=45)

# Add value annotations to the bars
for i, v in enumerate(winner_counts.values):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd

df_batting_summary = pd.read_excel('ICC_T20_WC_2022.xlsx', sheet_name='Batting_summary')

# Convert 'runs' and 'Notout' columns to numeric
df_batting_summary['runs'] = pd.to_numeric(df_batting_summary['runs'], errors='coerce')
df_batting_summary['Notout'] = pd.to_numeric(df_batting_summary['Notout'], errors='coerce')

# Calculate the total runs scored by each batsman
total_runs_by_batsman = df_batting_summary.groupby('batsmanName')['runs'].sum().sort_values(ascending=False)

number_of_times_notout = df_batting_summary.groupby('batsmanName').agg({'match': 'nunique', 'Notout': 'sum'})

# Calculate the number of times each player got out
number_of_times_out = number_of_times_notout['match'] - number_of_times_notout['Notout']

average = total_runs_by_batsman / number_of_times_out
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Filter the dataset to include relevant columns
df_filtered = df_bowling_summary[['bowlerName', 'runs', 'wickets', 'overs']]

# Convert 'overs' column to numeric
df_filtered['overs'] = pd.to_numeric(df_filtered['overs'], errors='coerce')

# Apply the condition of bowling at least 4 overs
df_filtered = df_filtered[df_filtered['overs'] >= 4]

# Calculate the bowling average for the entire tournament
df_filtered['bowling_average'] = df_filtered['runs'] / df_filtered['wickets']

# Ignore infinity values
df_filtered['bowling_average'] = df_filtered['bowling_average'].replace([np.inf, -np.inf], np.nan)

# Calculate the average for each player in the tournament
player_average = df_filtered.groupby('bowlerName')['bowling_average'].mean()

# Drop players with NaN average
player_average = player_average.dropna()

# Sort the players based on average in ascending order
top_20_bowlers = player_average.nsmallest(20)

# Plotting the top 20 bowlers with the best average
plt.figure(figsize=(12, 6))
bar_plot = plt.barh(top_20_bowlers.index, top_20_bowlers.values)
plt.xlabel('Bowling Average')
plt.ylabel('Bowler Name')
plt.title('Top 20 Bowlers with Best Bowling Average')
plt.gca().invert_yaxis()

# Display the values on the bars
for bar in bar_plot:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, '{:.2f}'.format(width), ha='left', va='center')

plt.show()

# Display the top 20 bowlers with the best average
print(top_20_bowlers)
```

```python
import pandas as pd
import numpy as np

# Filter the dataset to include relevant columns
df_filtered = df_bowling_summary[['bowlingTeam', 'runs', 'overs', 'Balls Bowled']]

# Convert 'overs' and 'Balls Bowled' columns to numeric
df_filtered['overs'] = pd.to_numeric(df_filtered['overs'], errors='coerce')
df_filtered['Balls Bowled'] = pd.to_numeric(df_filtered['Balls Bowled'], errors='coerce')

# Calculate the total number of balls bowled by each team
df_filtered['total_balls'] = df_filtered['overs'] * 6 + df_filtered['Balls Bowled']

# Calculate the total runs conceded by each team
team_runs = df_filtered.groupby('bowlingTeam')['runs'].sum()

# Calculate the total overs bowled by each team
team_overs = df_filtered.groupby('bowlingTeam')['overs'].sum()

# Calculate the bowling economy for each team
team_economy = team_runs / team_overs

# Ignore infinity values
team_economy = team_economy.replace([np.inf, -np.inf], np.nan)

# Sort the teams based on their bowling economy in ascending order
team_economy = team_economy.dropna().sort_values()

# Display the teams with their bowling economy
print(team_economy)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter the dataset to include relevant columns
df_filtered = df_bowling_summary[['match', 'bowlerName', 'wickets', 'runs']]

# Filter the dataset to include only the bowlers with 4 or more wickets in a match
four_wicket_bowlers = df_filtered[df_filtered['wickets'] >= 4]

# Sort the bowlers based on runs conceded in ascending order
sorted_bowlers = four_wicket_bowlers.sort_values(by='runs', ascending=True)

# Create separate lists for each occurrence of a 4-wicket haul
bowler_names = []
runs_conceded = []
wickets = []
for _, row in sorted_bowlers.iterrows():
    bowler_names.append(row['bowlerName'])
    runs_conceded.append(row['runs'])
    wickets.append(row['wickets'])

# Plotting the bowlers with 4-wicket hauls
plt.figure(figsize=(12, 8))
plt.bar(range(len(bowler_names)), runs_conceded, color='skyblue')

# Display the number of wickets on top of the respective bar
for i, (bowler, runs, wkt) in enumerate(zip(bowler_names, runs_conceded, wickets)):
    plt.text(i, runs, f'{wkt}/{runs}', color='black', ha='center', va='bottom', fontsize=8)

plt.xlabel('Bowler Name')
plt.ylabel('Runs Conceded')
plt.title('Bowlers with 4-Wicket Hauls')
plt.xticks(range(len(bowler_names)), bowler_names, rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
